```python
# Mengatur ukuran gambar (figure) dengan lebar 14 inci dan tinggi 6 inci.
plt.figure(figsize=(14, 6))

x = np.arange(len(data["Bulan"])) # Mengubah keterangan array x menjadi urutan angka dari 0 hingga panjang data kolom "Bulan".
Bulan = [str(i + 1) for i in range(len(data["Bulan"]))]
width = 0.28

# Membuat diagram batang dengan menggunakan data dari kolom '1981-1990', '1991-2000', dan '2001-2010'
plt.bar(x - 0.28, data["1981-1990"], color="#ADD8E6", edgecolor="k", width=width, label="1981-1990")
plt.bar(x + 0.00, data["1991-2000"], color="#FFC0CB", edgecolor="k", width=width, label="1991-2000")
plt.bar(x + 0.28, data["2001-2010"], color="#90EE90", edgecolor="k", width=width, label="2001-2010")

plt.xlabel("Bulan", fontsize=16, weight="bold")  # Menambahkan keterangan pada sumbu x dengan teks "Bulan"
plt.ylabel("Curah Hujan (mm)", fontsize=16, weight="bold")  # Menambahkan keterangan pada sumbu y dengan teks "Curah Hujan (mm)"

plt.ylim(0, 700)  # Menambahkan batasan curah hujan pada sumbu y menjadi 700

plt.legend(loc="upper center", ncol=3)  # Menampilkan legenda di bagian atas tengah gambar dengan 3 kolom (label untuk setiap periode)
plt.xticks(x, Bulan)  # Mengatur label sumbu x menggunakan nama bulan
plt.grid(zorder=0, color="gray", linestyle="dashed", linewidth=0.5)  # Menampilkan kisi-kisi dengan garis putus-putus pada latar belakang gambar

plt.show()  # Menampilkan gambar
```

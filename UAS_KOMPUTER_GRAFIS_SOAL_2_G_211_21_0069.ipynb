```python
data = pd.read_csv("SuhuPermukaan.csv")
```

```python
# Membuat array tanggal dan waktu dari tanggal '2013-07-24' dengan Frekuensi interval setiap 3 jam.
dates = pd.date_range(start='2013-07-24', periods=9, freq='3H')

plt.figure(figsize=(12, 6)) # Mengatur ukuran gambar (figure) dengan lebar 12 inci dan tinggi 6 inci.

labels = ['Asimilasi', 'Observasi', 'Non Asimilasi'] # Melakukan inisialisasi label dan warna untuk memplot data dari kolom yang sesuai dalam dataset.
colors = ['#800000', '#000080', '#00FF00']

for label, color in zip(labels, colors): # Melakukan perulangan data dari setiap kolom dalam dataset dengan menggunakan warna dan label yang sesuai.
  plt.plot(dates, data[label], marker='o', color=color, label=label)

for d in [20, 21, 22, 23, 24, 25, 26, 27]: # Menetapkan tinggi untuk setiap nilai y dalam plot hingga mencapai 27.
  plt.axhline(y=d, color='gray', linestyle='-', linewidth=0.3)

plt.gca().xaxis.set_major_formatter(dts.DateFormatter('%H:%M UTC')) # Menyesuaikan format sumbu x agar menampilkan jam dan menentukan penempatan utama pada sumbu x secara otomatis.
plt.gca().xaxis.set_major_locator(dts.AutoDateLocator())

plt.title('Suhu Permukaan', fontsize=16, weight='bold') # Memberi keterangan pada setiap label
plt.xlabel('Waktu', fontsize=14, weight='bold')
plt.ylabel('Celcius', fontsize=14, weight='bold')
plt.legend(loc='lower right')

# Menambahkan keterangan tanggal pada sumbu x
plt.text(dates[2], 19, '24 Jul 2013', ha='left', va='center')
plt.text(dates[-1], 19, '25 Jul 2013', ha='right', va='center')

plt.show()# Menampilkan gambar
```
